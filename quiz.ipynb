```python

# SMART QUIZ SYSTEM


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import time


# QUESTION  (20 General Questions)


questions = [
    {"category":"General Knowledge","question":"What is the capital of India?","options":["Mumbai","New Delhi","Kolkata","Chennai"],"answer":"New Delhi"},
    {"category":"General Knowledge","question":"Who is known as the Father of the Indian Constitution?","options":["Mahatma Gandhi","B. R. Ambedkar","Jawaharlal Nehru","Sardar Patel"],"answer":"B. R. Ambedkar"},
    {"category":"General Knowledge","question":"Which planet is known as the Red Planet?","options":["Earth","Mars","Venus","Jupiter"],"answer":"Mars"},
    {"category":"General Knowledge","question":"What is the largest ocean in the world?","options":["Indian Ocean","Pacific Ocean","Atlantic Ocean","Arctic Ocean"],"answer":"Pacific Ocean"},
    {"category":"General Knowledge","question":"Who wrote the national anthem of India?","options":["Rabindranath Tagore","Bankim Chandra Chatterjee","Sarojini Naidu","Subhash Chandra Bose"],"answer":"Rabindranath Tagore"},
    {"category":"General Knowledge","question":"Which gas do plants absorb from the atmosphere?","options":["Oxygen","Nitrogen","Carbon Dioxide","Hydrogen"],"answer":"Carbon Dioxide"},
    {"category":"General Knowledge","question":"Which is the longest river in the world?","options":["Ganga","Amazon","Nile","Yangtze"],"answer":"Nile"},
    {"category":"General Knowledge","question":"Which country is known as the Land of the Rising Sun?","options":["China","Thailand","Japan","South Korea"],"answer":"Japan"},
    {"category":"General Knowledge","question":"Who invented the telephone?","options":["Thomas Edison","Alexander Graham Bell","Newton","Tesla"],"answer":"Alexander Graham Bell"},
    {"category":"General Knowledge","question":"What is the chemical symbol of Gold?","options":["Go","Au","Ag","Gd"],"answer":"Au"},
    {"category":"General Knowledge","question":"Which continent is the Sahara Desert located in?","options":["Asia","Africa","Australia","Europe"],"answer":"Africa"},
    {"category":"General Knowledge","question":"Who was the first President of India?","options":["Dr. Rajendra Prasad","Dr. APJ Abdul Kalam","Jawaharlal Nehru","Indira Gandhi"],"answer":"Dr. Rajendra Prasad"},
    {"category":"General Knowledge","question":"Which is the smallest planet in our solar system?","options":["Mars","Mercury","Venus","Pluto"],"answer":"Mercury"},
    {"category":"General Knowledge","question":"What is the national animal of India?","options":["Lion","Elephant","Tiger","Leopard"],"answer":"Tiger"},
    {"category":"General Knowledge","question":"Which is the largest continent?","options":["Africa","Europe","Asia","North America"],"answer":"Asia"},
    {"category":"General Knowledge","question":"What is the boiling point of water?","options":["90°C","95°C","100°C","120°C"],"answer":"100°C"},
    {"category":"General Knowledge","question":"Who discovered gravity?","options":["Albert Einstein","Galileo","Isaac Newton","Nikola Tesla"],"answer":"Isaac Newton"},
    {"category":"General Knowledge","question":"Which is widely regarded as the national sport of India?","options":["Cricket","Hockey","Kabaddi","Football"],"answer":"Hockey"},
    {"category":"General Knowledge","question":"Which is the fastest land animal?","options":["Lion","Cheetah","Tiger","Horse"],"answer":"Cheetah"},
    {"category":"General Knowledge","question":"Which festival is known as the Festival of Lights?","options":["Holi","Diwali","Eid","Christmas"],"answer":"Diwali"}
]


# QUIZ FUNCTION

def start_quiz():
    random.shuffle(questions)
    score = 0
    results = []
    start_time = time.time()

    print("\n SMART QUIZ SYSTEM")
    print("----------------------------------")

    for i, q in enumerate(questions, 1):
        print(f"\nQuestion {i}: {q['question']}")
        for idx, option in enumerate(q["options"], 1):
            print(f"{idx}. {option}")

        try:
            choice = int(input("Enter option number: "))
            selected = q["options"][choice-1]
        except:
            selected = "Invalid"

        correct = selected == q["answer"]
        if correct:
            score += 1

        results.append(correct)

    total_time = round(time.time() - start_time, 2)

    print("\n✅ Quiz Completed!")
    print("Score:", score, "/", len(questions))
    print("Time Taken:", total_time, "seconds")

    return results

# ANALYSIS


def analyze_results(results):

    correct_count = sum(results)
    wrong_count = len(results) - correct_count

    print("\n PERFORMANCE SUMMARY")
    print("Correct Answers:", correct_count)
    print("Wrong Answers:", wrong_count)

    # Single Bar Graph
    plt.figure()
    plt.bar(["Correct", "Wrong"], [correct_count, wrong_count])
    plt.title("Quiz Result Analysis")
    plt.yticks(range(0,21 ))
    plt.ylabel("Number of Questions")
    plt.show()

# MAIN


if __name__ == "__main__":
    result_data = start_quiz()
    analyze_results(result_data)
```
